## Recorrer una API paginada y almacenar todos los datos en un solo dataframe

```python
import requests
import json
import pandas as pd
import time
from multiprocessing import Pool as ThreadPool
from multiprocessing import cpu_count

# offset=0 #Parametros iniciales
limit=100 # Parametrod iniciales
lista_offsets = [x*100 for x in range(0,30)]

def api_function(offset):
  print('|{}|'.format(offset))
  url="http://jsonplaceholder.typicode.com/photos?_start={}&_limit={}".format(offset,limit)
  result = requests.get(url)
  df=pd.json_normalize(result.json())
  return df
```

```python
s = time.time()
pool = ThreadPool(cpu_count())
results = pool.map(api_function, lista_offsets)
df = pd.concat(results)
df
print(time.time()-s)
```
