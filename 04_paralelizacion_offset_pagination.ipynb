## Recorrer una API paginada y almacenar todos los datos en un solo dataframe

```python
import requests
import json
import pandas as pd
import time
from multiprocessing import Pool as ThreadPool
from multiprocessing import cpu_count

offset=0 #Parametros iniciales
limit=1000 # Parametrod iniciales
lista_offset = [0,1000,2000,3000]

def api_function(lista_offset):
  print(lista_offset)
  lista=[]

  for offset in lista_offset:
    print(offset)
    url="http://jsonplaceholder.typicode.com/photos?_start={}&_limit={}".format(offset,limit)
    result = requests.get(url)
    data_extracted=result.json()
    lista.append(pd.json_normalize(data_extracted))
  return lista
```

```python
s = time.time()
pool = ThreadPool(4)
results = pool.map(api_function, [lista_offset])
df = pd.concat(results[0])
df
```
